Статистика по доменам базы последовательностей CAZy

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caz_data = pd.read_csv('/home/aster/Project_Geoarchaeota_28.11/results/dbCAN/cazy_data.txt', sep='\t', header=None)
```

```python
caz_data_parsed = caz_data[[0,1]]
```

```python
caz_data_parsed[0] = caz_data_parsed[0].apply(lambda x: ''.join([i for i in x if i.isalpha()]))
```

```python
caz_data_parsed
```

```python
vals = [len(caz_data_parsed[caz_data_parsed[1] == 'Archaea']),
        len(caz_data_parsed[caz_data_parsed[1] == 'Bacteria']),
        len(caz_data_parsed[caz_data_parsed[1] == 'Eukaryota']),
        len(caz_data_parsed[caz_data_parsed[1] == 'Viruses']),
        len(caz_data_parsed[caz_data_parsed[1] == 'unclassified'])]
```

```python
plt.figure(figsize=(10,10))
plt.pie(vals, labels=['Archaea','Bacteria','Eukaryota','Viruses','unclassified'])
plt.legend();
```
