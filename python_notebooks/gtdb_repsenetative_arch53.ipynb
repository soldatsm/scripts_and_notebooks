```python
import pandas as pd
```

```python
gtdb_table = pd.read_csv('/home/aster/Downloads/ar53_taxonomy_r207_reps.tsv', header=None, delimiter='\t')
```

```python
gtdb_table[['d', 'p', 'c','o','f','g', 's']] = gtdb_table[1].apply(lambda x: pd.Series(str(x).split(';')))
```

```python
gtdb_table[0] = gtdb_table[0].apply(lambda x: '_'.join(pd.Series(str(x).split('_')[1::])))
```

```python
gtdb_table = gtdb_table.drop(columns=[1])
```

```python
gtdb_table = gtdb_table.rename(columns={0:'Name', 'd':'domain', 'p': 'phylum', 'c': 'class', 'o':'order', 'f':'family','g':'genus', 's':'species'})
```

```python
gtdb_table.to_csv('/home/aster/Project_Geoarchaeota_28.11/gtdb207_ar53.tsv', sep='\t', index=False)
```
