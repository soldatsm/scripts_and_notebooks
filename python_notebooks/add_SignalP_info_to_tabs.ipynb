Добавление результатов программы SignalP в таблицу результатов геномного анализа.

```python
import pandas as pd
```

```python
signalp_geo3918 = pd.read_csv('../SignalIP/3918_out_dir/prediction_results.tsv', sep='\t',header=1)
data_tab_geo3918 = pd.read_csv('/home/aster/Downloads/Geo3918_МП_table_caz_merops.xlsx - Лист11.csv')
```

```python
signalp_geo3918['# ID'] = signalp_geo3918['# ID'].apply(lambda x: '_'.join(x.split('_')[0:2]))
```

```python
for idx, row in data_tab_geo3918.iterrows():
    for idx2, row2 in signalp_geo3918.iterrows():
        if row['pgap_loc_tag'] == row2['# ID']:
            data_tab_geo3918.loc[idx, 'SignalP_prediction'] = row2['Prediction']
            data_tab_geo3918.loc[idx, 'SignalP_CS_position'] = row2['CS Position']
```

```python
data_tab_geo3918.to_csv('signal_p_3918.csv')
```

```python
caz_data = pd.read_csv('/home/aster/Downloads/Geo3918_МП_table_caz_merops.xlsx - Лист12.csv')
```

```python
for idx, row in caz_data.iterrows():
    for idx2, row2 in signalp_geo3918.iterrows():
        if row['pgap_loc_tag'] == row2['# ID']:
            caz_data.loc[idx, 'SignalP_prediction'] = row2['Prediction']
            caz_data.loc[idx, 'SignalP_CS_position'] = row2['CS Position']
```

```python
caz_data.to_csv('signal_p_3918_caz.csv')
```
