## Batch faa processing

```python
path = './test_pgap/faas/'
for filename in os.listdir('./test_pgap/faas/'):
    
    fpath = os.path.join(path, filename)
    
    ids = []

    description = []

    for seq in SeqIO.parse(open(fpath, 'r'), 'fasta'):

        ids.append(seq.name)
        
        description.append(' '.join(seq.description.split('[')[0].split(' ')[1::]))

    df = pd.DataFrame({'ID':ids, 'Description':description})
    df.to_csv(f'{path}/{filename}.csv', index=False)    
```

## Batch variant

```python
import pandas as pd
import os
```

```python
path_to_folder = '../../pgap_annot/geo/data/'

for filename in os.listdir(path_to_folder):
    
    full_path = os.path.join(path_to_folder, filename)

    cds_list = []
    locus_tags = []#set()
    producsts = []
    prot_seq = []

    with open(full_path, 'r') as read_file:
        tmp_data = read_file.read()

    cds_list.append(tmp_data.split('     gene            ')[1::])
    
    for i in cds_list[0]:
        for j in  i.split('\n'):

            if '/locus_tag=' in j:

                proccesed_locus_tag = j.replace('                     /locus_tag="', '').strip('"')
                if proccesed_locus_tag in locus_tags:
                    continue
                else:
                    locus_tags.append(proccesed_locus_tag)
                
            if '/product="' in j:

                proccesed_prod = j.replace('                     /product="', '').strip('"')

                producsts.append(proccesed_prod)
        try:
            aa_seq = ''.join(i.split('/translation="')[1].replace('                     ','').replace('"', '').split('\n'))
            prot_seq.append(aa_seq.split(' ')[0])
        except IndexError:
            prot_seq.append('PSEUDO GENE')
            #prot_seq.append(''.join(i.split('/note="')[1].replace('                     ','').replace('"', '').split('\n')))

    data_frame = pd.DataFrame({'Locus_tag':locus_tags, 'Description':producsts, 'AA':prot_seq})  

    for idx, row in data_frame.iterrows():
        if row['AA'] != 'PSEUDO GENE':
            data_frame.loc[idx, 'Lenght'] = len(row['AA'])
        else:
            data_frame.loc[idx, 'Lenght'] =''
    data_frame.to_csv(f'{path_to_folder}/{filename}.csv', index=False)        
```
