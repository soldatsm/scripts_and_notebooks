## GC content GC_skew

```python
def CG_cont(lis:str) -> float:
    lis = lis.upper()
    GC = lis.count('G') + lis.count('C')
    return GC/len(lis)

def CG_skew(lis:str) -> float:
    lis = lis.upper()
    G = lis.count('G')
    C = lis.count('C')
    return (G - C)/(G + C)
```

```python
with open('/home/aster/Project_Geoarchaeota_28.11/data/sequences_annotation/1448933.3.contigs.fa', 'r') as read_file:
    genome_data = read_file.read().split('\n')[1::]
    genome_data = ''.join(genome_data)
```

```python
sliding_window = 5000
step = 0
step_size = 1000
r_bords = []
GCskew = []
GCcont = []
l_bords = []
for i in range(0,len(genome_data)):

    l_bord = step
    r_bord = sliding_window + l_bord
    step += step_size
    window = genome_data[l_bord:r_bord]

    l_bords.append(l_bord)
    r_bords.append(r_bord-sliding_window)
    GCskew.append(CG_skew(window))
    GCcont.append(CG_cont(window))

    if r_bord >= len(genome_data):
        break

GCskew.append((GCskew[0] + GCskew[-1])/2) #Замыкание кольца гистограммы
GCcont.append((GCcont[0] + GCcont[-1])/2)
l_bords.append(l_bords[-1] + sliding_window-step_size )
r_bords.append(r_bords[-1]  + sliding_window )
```

```python
for i,j,k, t in list(zip(r_bords, GCskew, GCcont, l_bords)):
    with open('/home/aster/Project_Geoarchaeota_28.11/results/circos_diagramms/data/GC_skew', 'a') as skew_file:
        if j > 0:
            skew_file.write(f"genome {t+1} {i+1} {j} fill_color=green \n")
        else:
            skew_file.write(f"genome {t+1} {i+1} {j} fill_color=red \n")

    with open('/home/aster/Project_Geoarchaeota_28.11/results/circos_diagramms/data/GC_cont', 'a') as cont_file:
            #k2 = (k - min(GCcont))/(max(GCcont)-(min(GCcont))) #minmax norm
            if k > 0.5:
                cont_file.write(f"genome {t+1} {i+1} {k} fill_color=blue\n")
            else:
                cont_file.write(f"genome {t+1} {i+1} {k} fill_color=orange\n")
```

## CG_skew

```python
def CG_cont(lis:str) -> float:
    lis = lis.upper()
    GC = lis.count('G') + lis.count('C')
    return GC/len(lis)

def CG_skew(lis:str) -> float:
    lis = lis.upper()
    G = lis.count('G')
    C = lis.count('C')
    return (G - C)/(G + C)
```

```python
def slider(seq,sliding_window, step_size):
    #sliding_window = 1000
    step = 0
    #step_size = 1000
    r_bords = []
    GCskew = []
    #GCcont = []
    l_bords = []
    for i in range(0,len(seq)):

        l_bord = step
        r_bord = sliding_window + l_bord
        step += step_size
        window = seq[l_bord:r_bord]

        l_bords.append(l_bord+1)
        r_bords.append(r_bord-sliding_window+1)
        GCskew.append(CG_skew(window))
        #GCcont.append(CG_cont(window))

        if r_bord >= len(seq):
            break

    GCskew.append((GCskew[0] + GCskew[-1])/2) #Замыкание кольца гистограммы
    #GCcont.append((GCcont[0] + GCcont[-1])/2)
    l_bords.append(l_bords[-1] + sliding_window-step_size )
    r_bords.append(r_bords[-1]  + sliding_window )
    return list(zip(l_bords,r_bords,GCskew))
```

```python
for sequence in SeqIO.parse(open('/home/aster/Trmosos_files/results/circos_diagramm/data/2893809466.fna','r'), 'fasta'):
    with open('/home/aster/Trmosos_files/results/circos_diagramm/data/A10_GC_skew.txt','a') as skew_file:

        if len(sequence.seq) >= 5000:
            lst = slider(str(sequence.seq), sliding_window=5000,step_size=1000)
            for i,j,k in lst:
                if k > 0:
                    skew_file.write(f"{sequence.name} {i} {j} {k} fill_color=green\n")
                else:
                    skew_file.write(f"{sequence.name} {i} {j} {k} fill_color=red\n")

        else:
            lst = slider(str(sequence.seq), sliding_window=1000,step_size=1000)
            for i,j,k in lst:
                if k > 0:
                    skew_file.write(f"{sequence.name} {i} {j} {k} fill_color=green\n")
                else:
                    skew_file.write(f"{sequence.name} {i} {j} {k} fill_color=red\n")
```
