## Добавление CAZy данных 
Есть две таблицы которы я качаю с dbCAN2:
* dbsub.out.tsv - Из этой таблицы стоит взять `dbCAN_subfam`, `Subfam_composition`, `substrate`, `E-Value`, `Coverage` 
* overview.tsv - `HMMER`, `DIMOND`, `Signalp`, `#ofTools`

```python
dbsub_df = pd.read_csv('/home/aster/Project_Geoarchaeota_28.11/3817-70/results/CAZy/dbsub.out.tsv', delimiter='\t')
overwiev_df = pd.read_csv('/home/aster/Project_Geoarchaeota_28.11/3817-70/results/CAZy/overview.tsv', delimiter='\t')
```

```python
dbsub_df.head()
```

```python
overwiev_df.head()
```

```python

# Добавление данных из dbsub
for idx1, row1 in genome_tab.iterrows():
    for idx2, row2 in dbsub_df.iterrows():
        if row1['IDs'] == row2['Gene ID']:
            genome_tab.loc[idx1, 'dbCAN subfam'] = row2['dbCAN subfam']
            genome_tab.loc[idx1, 'Subfam Composition'] = row2['Subfam Composition']
            genome_tab.loc[idx1, 'Subfam EC'] = row2['Subfam EC']
            genome_tab.loc[idx1, 'E Value'] = row2['E Value']
            genome_tab.loc[idx1, 'Substrate'] = row2['Substrate']
            genome_tab.loc[idx1, 'Coverage'] = row2['Coverage']
```

```python
# Добавление данных из overview
for idx1, row1 in genome_tab.iterrows():
    for idx2, row2 in overwiev_df.iterrows():
        if row1['IDs'] == row2['Gene ID']:
            genome_tab.loc[idx1, 'HMMER'] = row2['HMMER']
            genome_tab.loc[idx1, 'DIAMOND'] = row2['DIAMOND']
            genome_tab.loc[idx1, '#ofTools'] = row2['#ofTools']
```

```python
genome_tab.to_csv('/home/aster/Project_Geoarchaeota_28.11/3817-70/results/CAZy/Geo3817_CAZy_only.tsv', index=False)
```
