```python
from Bio import SeqIO
import pandas as pd
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

# Making tablets from fasta

Make table from thous fasta files obtained from GeneBank. 

reference my older script that make table (Maybe I should rewrite it in future.)

```python
# read fasts file:

path = '/home/aster/Project_Geoarchaeota_28.11/pgap_annot/other_arch/genomes/folder_with_proteomes/'
```

```python
tables_lst = []
for file in os.listdir(path):
    fpath = os.path.join(path, file)

    accesion = file.split('.')[0]
    pre_table = pd.DataFrame()

    for idx, seq in enumerate(SeqIO.parse(fpath, 'fasta')):
        
        
        protein_id = seq.description.split(' ')[0]

        protein_name = seq.description.split('[')[0].split(',')[0].split(' ')
        protein_name = ' '.join([i for i in protein_name if '.1' not in i and '_' not in i]) 

        locus_tag = seq.description.split('[')[0].split(',')[0].split(' ')
        locus_tag = ''.join([i.lstrip('[').rstrip(']') for i in locus_tag if '_' in i ])

        if len(locus_tag) == 0:
            locus_tag = np.NaN

        org_name = set()
        org_name.add(str(seq.description.split('[')[1].rstrip(']')))
        org_name = list(org_name)

        sequence = str(seq.seq)
        prot_len = len(sequence)

        pre_table.loc[idx, 'Gene id'] = protein_id
        pre_table.loc[idx, 'Name'] = protein_name
        pre_table.loc[idx, 'Locus Tag'] = locus_tag
        pre_table['Organism Name'] = org_name[0]
        pre_table.loc[idx, 'Protein lenght'] = prot_len
        pre_table.loc[idx, 'AAs'] = sequence
    pre_table['Accession'] = accesion

    tables_lst.append(pre_table)
```

First of all make a separate folder for all CAZy  data.

```python
for file in tqdm(os.listdir(caz_path)):
    fpath = os.path.join(caz_path, file)
    for table in tables_lst:
        #print(table.iloc[1,-1])
        if table.loc[0, 'Accession'] in file:
            print(file)
            if 'dbsub' in file:
                dbsub_tab = pd.read_csv(fpath, sep='\t')

                for idx, row in tqdm(table.iterrows()):
                    for idx2, row2 in dbsub_tab.iterrows():
                        if row['Gene id'] == row2['Gene ID']:
                            table.loc[idx, 'dbCAN subfam'] = row2['dbCAN subfam']
                            table.loc[idx, 'Subfam Composition'] = row2['Subfam Composition']
                            table.loc[idx, 'Subfam EC'] = row2['Subfam EC']
                            table.loc[idx, 'Substrate'] = row2['Substrate']
                            table.loc[idx, 'Coverage'] = row2['Coverage']
                            table.loc[idx, 'E Value'] = row2['E Value']

            if 'CAZ' in file:
                CAZ_over_tab = pd.read_csv(fpath, sep='\t')
                
                for idx, row in table.iterrows():
                    for idx2, row2 in CAZ_over_tab.iterrows():
                        if row['Gene id'] == row2['Gene ID']:
                            table.loc[idx, 'HMMER'] = row2['HMMER']
                            table.loc[idx, 'dbCAN_sub'] = row2['dbCAN_sub']
                            table.loc[idx, 'DIAMOND'] = row2['DIAMOND']
                            table.loc[idx, '#ofTools'] = row2['#ofTools']
```

### Gather a statistic data. For CAZy fam.

```python
def caz_gather(table:pd.DataFrame, conc=None):
    CAZ_lst = [i.split('_')[0] for i in table['dbCAN subfam'].dropna().to_list()]
    #CAZ_lst = [i.split('_')[0] for i in table['dbCAN_sub'].dropna().to_list()]

    GH = 0
    PL = 0
    CE = 0
    GT = 0
    AA = 0
    CBM = 0

    for i in CAZ_lst:
        if 'GH' in i:
            GH += 1
        elif 'PL' in i:
            PL += 1
        elif 'CE' in i:
            CE += 1
        elif 'GT' in i:
            GT += 1
        elif 'AA' in i:
            AA += 1
        elif 'CBM' in i:
            CBM += 1
    
    numbers = [('GH', GH), ('PL', PL), ('CE', CE), ('GT', GT), ('AA', AA), ('CBM', CBM)]

    agregate_table = pd.DataFrame({'GH':GH, 'PL':PL, 'CE':CE, 'GT':GT, 'AA':AA, 'CBM':CBM}, index=[0])

    test_lst = []
    for i in CAZ_lst:
        test_lst.append(CAZ_lst.count(i))

    all_table = pd.DataFrame({"CAZ":CAZ_lst, "Num":test_lst}).drop_duplicates(subset='CAZ')

    if conc is None:
        
        return agregate_table, all_table
    
    else:

        agregate_table_t = agregate_table.T
        agregate_table_t['CAZ'] = agregate_table_t.index
        agregate_table_t = agregate_table_t[['CAZ', 0]]
        agregate_table_t = agregate_table_t.rename({0:"Num"}, axis=1)
        
        df_conc = pd.concat([all_table, agregate_table_t], ignore_index=True)
        return df_conc
```
