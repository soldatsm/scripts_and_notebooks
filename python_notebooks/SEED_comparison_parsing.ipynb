```python
import pandas as pd
```

```python
SEED_compar_df = pd.read_csv('../SEED/Geo3918_toGeo3817_seq_compar.tsv', sep='\t')
```

```python
SEED_compar_df
```

```python
SEED_compar_df = SEED_compar_df.drop(columns=['Gene', 'Length', 'Gene.1'])
```

```python
all_tab_df_3918 = pd.read_csv('../SEED/Geo3918_МП_table_caz_merops.xlsx - Лист11.csv')
all_tab_df_3817 = pd.read_csv('../SEED/geo3817_AK_full_table - Лист7.csv')
```

```python
for idx, row in SEED_compar_df.iterrows():
    for idx2, row2 in all_tab_df_3918.iterrows():
        if row['Gene id'] == row2['Gene id']:
            SEED_compar_df.loc[idx, 'PGAP_ID_3918'] = row2['pgap_loc_tag']
            SEED_compar_df.loc[idx, 'PGAP_descriptiopn_3918'] = row2['pgap_descr']
            SEED_compar_df.loc[idx, 'AA_3918'] = row2['AAs']

for idx, row in SEED_compar_df.iterrows():
    for idx2, row2 in all_tab_df_3817.iterrows():
        if row['Gene id.1'] == row2['IDs']:
            SEED_compar_df.loc[idx, 'PGAP_ID_3817'] =  row2['pgap_loc_tag']
            SEED_compar_df.loc[idx, 'PGAP_descriptiopn_3817'] = row2['pgap_descr']
            SEED_compar_df.loc[idx, 'AA_3817'] = row2['AAs']
```

```python
SEED_compar_df
```

```python
SEED_compar_df = SEED_compar_df[['Contig', 'Gene id', 'AA_3918','function','PGAP_ID_3918','PGAP_descriptiopn_3918',  'Contig.1', 'Gene id.1',
       'percent identity 1448933.6','AA_3817', 'function.1', 'PGAP_ID_3817', 'PGAP_descriptiopn_3817', 'Hit']]
```

```python
SEED_compar_df['percent identity 1448933.6'] = SEED_compar_df['percent identity 1448933.6'].apply(lambda x : ','.join(str(x).split('.')))
```

```python
SEED_compar_df.to_csv('../SEED/Geo_SEED_comparison.tsv', index=False, sep='\t')
```
