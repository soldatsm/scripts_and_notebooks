```python
def bone_devil(tab1:pd.core.frame.DataFrame, tab2:pd.core.frame.DataFrame, prot_name_idx, pgap_idx, pgap_desc_idx, output):
    for idx, row in tab1.iterrows():
        for idx2, row2 in tab2.iterrows():
            if row.iloc[prot_name_idx] ==  row2.iloc[prot_name_idx]:
                tab2.loc[idx2, 'pgap_loc_tag'] = row.iloc[pgap_idx]
                tab2.loc[idx2, 'pgap_descr	'] = row.iloc[pgap_desc_idx]
    tab2.to_csv(f'{output}', index=False)
```

```python
bone_devil(pgap_3817, caz_3817, 0, 1, 3, './caz_pgap_3817.csv')
bone_devil(pgap_3817, pept_3817, 0, 1, 3, './pep_pgap_3817.csv')
bone_devil(pgap_3817, tm_3817, 0, 1, 3, './tm_pgap_3817.csv')
```

```python
caz_3918.iloc[:, 0] = caz_3918.iloc[:, 0].apply(lambda x:x.split('|')[1])
pept_3918.iloc[:, 0] = pept_3918.iloc[:, 0].apply(lambda x:x.split('|')[1])
tm_3918.iloc[:, 0] = tm_3918.iloc[:, 0].apply(lambda x:x.split('|')[1])
```
