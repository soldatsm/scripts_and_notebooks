## Сделать фасту для обратного бласта

Проверка будет простая, просто обратный бласт на базу swiss-prot, для этого нужно сделать по хитам соответвующий фаста файл.

```python
#Чтение файла с конкатенированным протеомом
c = 0
for seq in SeqIO.parse(open('/home/aster/Project_Geoarchaeota_28.11/results/mars_comarison/blast_db_concat_Mars_all/mars_concatenated.faa'), 'fasta'):
    for hit in data.itertuples():
        if hit[2] in seq.name:
            c+=1
            print(hit[2])
print(c)
```

Колличество совпадает так что все работает нормально

```python
with open('./mer.besthits.fasta', 'a') as write_file:
    for seq in SeqIO.parse(open('/home/aster/Project_Geoarchaeota_28.11/results/mars_comarison/blast_db_concat_Mars_all/mars_concatenated.faa'), 'fasta'):
        for hit in data.itertuples():
            if hit[2] in seq.name:
                write_file.write(f">{hit[2]}\n{seq.seq}\n")
```
