## 3918

```python
import pandas as pd
```

```python
overwiew =  pd.read_csv('/home/aster/Project_Geoarchaeota_28.11/pgap_annot/geo/3918_pgap_rast_CAZ_overview.txt', sep='\t')
dbsub = pd.read_csv('/home/aster/Project_Geoarchaeota_28.11/pgap_annot/geo/3918_pgap_rast_CAZ_dbsub.txt', sep='\t')
data = pd.read_csv('./fixed_pgap_geo3918_sorted.csv')
```

```python
dbsub.head(2)
```

```python
overwiew.head(2)
```

```python
data.head(2)
```

```python
dbsub.head(1)
```

```python
for idx, row in data.iterrows():
    for idx2, row2 in overwiew.iterrows():
        if row['pgap_loc_tag'] == row2['Gene ID'] or row['Gene id'] == row2['Gene ID']:
            data.loc[idx, 'EC#'] = row2['EC#']
            data.loc[idx, 'HMMER'] = row2['HMMER']
            data.loc[idx, 'dbCAN_sub'] = row2['dbCAN_sub']
            data.loc[idx, 'Signalp'] = row2['Signalp']
            data.loc[idx, '#ofTools'] = row2['#ofTools']
            data.loc[idx, 'DIAMOND'] = row2['DIAMOND']
    for idx3, row3 in dbsub.iterrows():
        if row['pgap_loc_tag'] == row3['Gene ID'] or row['Gene id'] == row3['Gene ID']:
            data.loc[idx, 'dbCAN subfam/MEROPS'] = row3['dbCAN subfam']
            data.loc[idx, 'Substrate/description'] = row3['Substrate']
            data.loc[idx, 'E Value'] = row3['E Value']
            data.loc[idx, 'Coverage'] = row3['Coverage']
            data.loc[idx, 'Subfam Composition'] = row3['Subfam Composition']
            #data.loc[idx, 'DIAMOND'] = row2['DIAMOND']
```

```python
d2 = data.dropna(subset='dbCAN subfam/MEROPS')
d2[~d2['dbCAN subfam/MEROPS'].str.contains('M')]
caz_3918 = d2[~d2['dbCAN subfam/MEROPS'].str.contains('M')]
```

```python
data.to_csv('./fixed_pgap_geo3918+CAZ_sorted.csv', index=False)
caz_3918.to_csv('./rast_pgap_CAZ_only.csv', index=False)
```
