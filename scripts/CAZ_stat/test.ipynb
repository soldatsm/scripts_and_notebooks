```python
dbsub_parsing = dbsub[['dbCAN subfam', 'Subfam Composition']]
```

```python
str.isnumeric
```

```python
for idx, row in dbsub_parsing.iterrows():

    parsed_type = [i for i in list(row['dbCAN subfam']) if i.isalpha() and i != 'e']

    dbsub_parsing.loc[idx, 'CAZ_type'] = ''.join(parsed_type)
```

```python
dbsub_parsing
```

```python
dbsub_parsing
```

```python
# Count Cazyms
count_CAZ_type_set = set()
caz_lst = dbsub_parsing['CAZ_type'].to_list()
for i in caz_lst:
    count_CAZ_type_set.add((i, caz_lst.count(i)))

#Coun different cazyms
count_CAZ_all_set = set()
caz_lst = dbsub_parsing['dbCAN subfam'].to_list()
for i in caz_lst:
    count_CAZ_all_set.add((i, caz_lst.count(i)))
```

```python
count_CAZ_type_set = sorted(list(count_CAZ_type_set), key=lambda x: x[0]) # Сортировка по именам, по идеи этоь делает все это робастным
count_CAZ_all_set = sorted(list(count_CAZ_all_set), key=lambda x: x[0])
```

```python
plt.figure(figsize=(10,10), )

#count_CAZ_type_set
#count_CAZ_all_set

plt.pie([i[1] for i in count_CAZ_type_set], labels=[f'{i[0]}({i[1]})' for i in count_CAZ_type_set], 
        radius=0.5, wedgeprops=dict(width=0.5, edgecolor='w'), labeldistance=0.5, textprops={'fontsize': 10}, rotatelabels=270)

plt.pie([i[1] for i in count_CAZ_all_set], labels=[f'{i[0]} ({i[1]})' for i in count_CAZ_all_set], 
        radius=1, wedgeprops=dict(width=0.5, edgecolor='w'),rotatelabels=270, labeldistance=0.6, textprops={'fontsize': 9} )


plt.title('Carboxy active enzymes pie chart for {Genome}', fontsize=20)
plt.tight_layout()
#plt.show()
```

```python
def drawer(list_types:list[tuple], list_all:list[tuple], genome_name:str):
    
    plt.figure(figsize=(10,10), )

    #count_CAZ_type_set
    #count_CAZ_all_set

    plt.pie([i[1] for i in list_types], labels=[f'{i[0]}({i[1]})' for i in list_types], 
            radius=0.5, wedgeprops=dict(width=0.5, edgecolor='w'), labeldistance=0.5, textprops={'fontsize': 10}, rotatelabels=270)

    plt.pie([i[1] for i in list_all], labels=[f'{i[0]} ({i[1]})' for i in list_all], 
            radius=1, wedgeprops=dict(width=0.5, edgecolor='w'),rotatelabels=270, labeldistance=0.6, textprops={'fontsize': 9} )


    plt.title(f'Carboxy active enzymes pie chart for {genome_name}', fontsize=20)
    plt.tight_layout()
```

```python
drawer(count_CAZ_type_set, count_CAZ_all_set, 'test')
```
